```python
import torch
import torch.nn as nn
import torch.nn.functional as F

class SelfAttention(nn.Module):
    def __init__(self, embed_size, head_size):
        # Argument:
        #   embed_size (int): The size of the embedding dimension.
        #   head_size (int): The size of the query/key/value vector dimension.
        # Submodules:
        #   self.query (nn.Linear): Linear transformation for the query vectors.
        #   self.key (nn.Linear): Linear transformation for the key vectors.
        #   self.value (nn.Linear): Linear transformation for the value vectors.

        super(SelfAttention, self).__init__()
        self.embed_size = embed_size
        self.head_size = head_size

        self.query = nn.Linear(embed_size, head_size)
        self.key = nn.Linear(embed_size, head_size)
        self.value = nn.Linear(embed_size, head_size)

    def forward(self, x):
        """
        Forward pass for SelfAttention.

        Args:
            x (Tensor): Input tensor of shape (batch_size, seq_length, embed_size).

        Returns:
            Tensor: The output after single-head attention.
        """

        output = None

        #### FILL YOUR CODE BELOW ###
        # Project x to query/key/value vectors;
        # HINT: 3 lines
        query = self.query(x)
        key = self.key(x)
        value = self.value(x)
        #############################

        #### FILL YOUR CODE BELOW ###
        # Compute attention scores,
        # notice that the raw scores will be divided by \sqrt{self.embed_size}
        # HINT: you can either implement using
        # 1. torch.bmm() to compute the attention score for a batch in parallel, this requires one line of code
        # 2. Use for-loop to iterate over each sequence in the batch for computing the scores, this may require more lines of code
        scores = torch.bmm(query, key.transpose(1, 2)) / torch.sqrt(torch.tensor(self.embed_size))
        #############################

        #### FILL YOUR CODE BELOW ###
        # Apply softmax to compute attention weights
        attention_weights = torch.softmax(scores, dim = -1)
        #############################

        #### FILL YOUR CODE BELOW ###
        # Use attn_weights to perform Weighted sum on value
        output = torch.bmm(attention_weights, value)
        #############################

        return output
```

Implement the `LayerNormalization` module. To simplify the implementation, we **DO NOT** employ dropout as the original BERT paper.

```python
class LayerNormalization(nn.Module):
    def __init__(self, embed_size, eps=1e-6):
        """
        Layer Normalization normalizes across the feature dimension (last dimension).
        It ensures that the mean and variance of the output remain constant and
        adds learnable scale and shift parameters.

        Args:
            embed_size (int): The size of the embedding dimension.
            eps (float): A small value to avoid division by zero during normalization.
        """
        super(LayerNormalization, self).__init__()
        self.embed_size = embed_size
        self.eps = eps

        #### FILL YOUR CODE BELOW ####
        # Define learnable parameters gamma and beta
        # Note: Initialize `gamma` with an all-one vector.
        # Note: Initialize `beta` with an all-zero vector.
        self.gamma = torch.ones(embed_size)
        self.beta = torch.zeros(embed_size)
        ##############################

    def forward(self, x):
        """
        Forward pass for LayerNormalization.

        Args:
            x (Tensor): Input tensor of shape (batch_size, seq_length, embed_size).

        Returns:
            Tensor: The normalized output with the same shape as the input.
        """
        output = None

        #### FILL YOUR CODE BELOW ####
        # Compute mean and variance across the last dimension
        # HINT: Use torch.var to calculate variance and remember to set unbiased=False
        mean = x.mean(dim = -1, keepdim = True)
        variance = torch.var(x, dim = -1, unbiased = False, keepdim = True)
        ##############################

        #### FILL YOUR CODE BELOW ####
        # Normalize the input
        norm_x = (x - mean) / torch.sqrt(variance + self.eps)
        ##############################

        #### FILL YOUR CODE BELOW ####
        # Apply gamma (scale) and beta (shift)
        output = self.gamma * norm_x + self.beta
        ##############################

        return output
```
